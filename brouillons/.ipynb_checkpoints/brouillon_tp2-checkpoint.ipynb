```python
import numpy as np
np.set_printoptions(threshold=10000,suppress=True)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

## Load dataset

```python
data = pd.read_csv('./villes.csv', sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values
print(labels)

data.head()
print(data.info())
```

## Preprocess dataset
bring all features to a common scale (centré réduit)

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
scaled_data = std_scaler.fit_transform(X)

print("vérifier que la moyenne des colonnes est nulle :")
print(scaled_data.mean(axis = 0))

#std: standard deviation = écart type
print("vérifier que l'écart type des colonnes est 1 :")
print(scaled_data.std(axis = 0))
```

## Perform PCA on preprocessed data
on veut réduire la dimension des données, passer de 12 à la plus petite dimension qui nous permet de garder 90% de la représentation de l'information initiale

**Note : Le ratio de variance expliquée mesure quelle proportion de la variance totale est capturée par chaque composante principale.**

```python
from sklearn.decomposition import PCA

var_ratio = []
for i in range(1,13):
    pca = PCA(n_components = i)
    pca.fit_transform(scaled_data)
    #on remarque ici que lorsqu'on rajoute une composante à l'ACP le ratio de variance de cette nouvelle variable est très petit (elle ne rajoute pas trop d'info à ce qu'on avait déjà)
    print(pca.explained_variance_ratio_)
    var_ratio.append(np.sum(pca.explained_variance_ratio_))

var_ratio = np.array(var_ratio)

nb_axes_min_90 = np.argmax(var_ratio > 0.9) + 1 
print(nb_axes_min_90)
```

le nombre d'axes min à retenir pour conserver un minimum de 90% de l'info initiale est 2


```python
#visualisation des ratios de variances selon le nombre de composants
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(np.arange(1,13),var_ratio,marker='o')
plt.xlabel('n_components')
plt.ylabel('Explained variance ratio')
plt.title('n_components vs. Explained Variance Ratio')

plt.show()
```

**On remarque qu'à partir de 2 composants on atteint 90%**

- approche différente .. pas trop
on fait directement l'ACP pour 12 composants et la somme cumulative donnera le ratio de variacne pour chaque x composants...
on peut déterminer comme ça combien de composants min il nous faut pour garder 90% de l'info initiale
au lieu de faire une acp à chaque fois
- var_ratio contient : ratio 1 composant , ratio 2 composants, ratio .....

```python
pca12 = PCA(n_components = 12)
pca12.fit_transform(scaled_data)
var_ratio = np.cumsum(pca12.explained_variance_ratio_)
print(var_ratio)

nb_min_axes_90 = np.argmax(var_ratio > 0.9) + 1;
print(nb_min_axes_90)
```

## c) Visualisation graphique des villes projetées dans le plan principal

principe: le produit de la matrice initiale (X_data) et la matrice composée des deux vecteur composants principaux nous donne une matrice d'observation à projeter sur un plan 2D

ou bien ... : X_pca est la matrice des données transformée par l'ACP (fit_transform)

```python
pca_2d = PCA(n_components = 2)
X_pca = pca_2d.fit_transform(scaled_data)
print(X_pca)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1])
for l, x, y in zip(labels, X_pca[:, 0], X_pca[:, 1]):
    plt.annotate(l, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')

plt.grid()
plt.xlabel("Premier axe principal")
plt.ylabel("Deuxième axe principal")
plt.title("Projection des villes dans le plan principal")

plt.show()
```

## Définir une fonction permettant de regrouper toutes les procédures précédentes

```python
def analyse_acp(filename_csv):
    #load data in dataframe
    data = pd.read_csv(filename_csv, sep=';')
    maxCol = data.shape[1]  # Get the total number of columns
    X = data.iloc[:, 1:maxCol].values
    labels = data.iloc[:, 0].values

    #preprocess data
    std_scaler = StandardScaler()
    scaled_data = std_scaler.fit_transform(X)

    #apply pca and find the best number of principal components to get 90% of initial data
    pca_max = PCA(n_components = maxCol - 1)
    pca_max.fit_transform(scaled_data)
    var_ratio = np.cumsum(pca_max.explained_variance_ratio_)

    print(var_ratio)
    nb_min_axes_90 = np.argmax(var_ratio > 0.9) + 1;
    print(f"Minimum number of components to explain 90% variance: {nb_min_axes_90}")

    #apply pca with nb_max_axis_90 components
    pca = PCA(n_components = nb_min_axes_90)
    X_pca = pca.fit_transform(scaled_data)
    print(X_pca)
    #to visualize data we'll choose the 2 first PCs
    plt.scatter(X_pca[:, 0], X_pca[:, 1])
    for l, x, y in zip(labels, X_pca[:, 0], X_pca[:, 1]):
        plt.annotate(l, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
    
    plt.grid()
    plt.xlabel("Premier axe principal")
    plt.ylabel("Deuxième axe principal")
    plt.title("Projection des données itiales dans le plan principal")
    
    plt.show()
```
