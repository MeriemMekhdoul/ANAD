```python
import numpy as np
np.set_printoptions(threshold=10000,suppress=True)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

## 1. Importer le jeu de données

Le fichier "villes.csv" comporte 32 villes françaises décrites par les températures moyennes dans les 12 mois de
l’année. L’objectif dans cette partie est de représenter graphiquement le plus d’informations possibles contenues
dans ce fichier de données et de déceler une éventuelle segmentation topologique des villes.

```python
data = pd.read_csv('./villes.csv', sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values

data.head()
```

## 2. Réaliser une ACP (Analyse en Composantes Principales)

### la matrice de données doit être centrée réduite
- Utilisation de la bibliothèque Scikit-Learn

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

#vérification
print(scaled_data.mean(axis =0))
print(scaled_data.std(axis =0))
```

### a) Trouver le nombre min d'axes à retenir pour conserver 90% de l'information initiale
principe : 
- on fait une acp en x composants (x allant de 1 à 12, nombres de colonnes max de la matrice initiale) (on peut aussi s'arrêter dès qu'on atteint le taux qui nous intéresse)
- pour chaque acp on garde le ratio de variance dans un tableau **var_ratio**
- l'attribut **explained_variance_ratio** capture le ratio de la variance totale capturée par chaque composante principale. Nous pouvons donc additionner les ratios pour obtenir la variance totale du nombre de composantes choisi. càd qu'il montre pour chaque axe à quel point il est représentatif des données initiales.
- pour trouver le nombre d'axes min pour garantir une conservation de 90% de l'info on prend le plus petit indice du tabelau qui contient une valeur > 0.9

```python
from sklearn.decomposition import PCA
var_ratio = []

for i in range(1,13):
    pca = PCA(n_components = i)
    X_pca = pca.fit_transform(scaled_data)
    var_ratio.append(sum(pca.explained_variance_ratio_))
```

```python
var_ratio = np.array(var_ratio)
nb_axes_90 = np.argmax(var_ratio > 0.9) + 1

print("Le nombre minimum d'axes à retenir pour conserver 90% de l'information initiale est :",nb_axes_90)
```

### b) Donner une interprétation des deux premiers axes principaux

Le **cercle de corrélation ci-dessous** illustre la contribution de chaque variable principale à la construction des nouveaux axes.  

L’axe PC1 est une combinaison linéaire de toutes les variables initiales, avec des coefficients de contribution équivalents. Cela signifie que **PC1 représente la moyenne annuelle des températures des différentes villes.**  

Concernant l’axe PC2, on observe que les mois de janvier, février, novembre et décembre ont un coefficient supérieur à 0.25, tandis que mai, juin, juillet et août affichent un coefficient inférieur à -0.25. Les mois d’hiver sont donc fortement corrélés à PC2, tandis que les mois d’été le sont inversement.  

Ainsi, **PC2 reflète l’écart entre les températures hivernales et estivales** : plus sa valeur est élevée, plus les températures hivernales sont élevées et les températures en été le sont moins; à l’inverse, plus elle est basse, plus les températures en été sont élevées et en hivers plus faibles.

```python
#fonction pour affiche le cercle de corrélation des variables initiales sur les nouveaux axes
def plot_correlation_circle(pca, feature_names, pc_x=0, pc_y=1):
    """
    Trace le cercle des corrélations pour une paire de composantes principales.
    
    Arguments :
    - pca : objet PCA déjà ajusté
    - feature_names : liste des noms des variables initiales
    - pc_x : indice de la première composante principale
    - pc_y : indice de la deuxième composante principale
    """
    loadings = pca.components_.T  # Charge les coefficients des variables
    fig, ax = plt.subplots(figsize=(4,4))

    # Cercle unité
    circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='dashed', alpha=0.5)
    ax.add_patch(circle)
    
    # Flèches représentant les variables sur les axes PC_x et PC_y
    for i, (x, y) in enumerate(zip(loadings[:, pc_x], loadings[:, pc_y])):
        ax.arrow(0, 0, x, y, head_width=0.05, head_length=0.05, fc='red', ec='red')
        ax.text(x, y, feature_names[i], fontsize=12, ha='right' if x < 0 else 'left', color='black')

    # Axes et limites
    ax.axhline(0, color='gray', linestyle='dashed', linewidth=0.5)
    ax.axvline(0, color='gray', linestyle='dashed', linewidth=0.5)
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.set_xlabel(f"Composante principale {pc_x+1}")
    ax.set_ylabel(f"Composante principale {pc_y+1}")
    ax.set_title(f"Cercle des Corrélations (PC{pc_x+1} vs PC{pc_y+1})")
    plt.grid()

    plt.show()
```

```python
pc1 = pca.components_[0]
pc2 = pca.components_[1]

print(pc1)
print(pc2)

plot_correlation_circle(pca.fit(scaled_data), data.columns[1:13], pc_x=0, pc_y=1)
```

### e) Définir une fonction permettant de regrouper toutes les procédures précédentes 


```python
import seaborn as sns
def analyse_acp(filename_csv):
    # Charger les données
    data = pd.read_csv(filename_csv, sep=';')
    maxCol = data.shape[1]  # Nombre total de colonnes
    X = data.iloc[:, 1:maxCol].values  # Données numériques
    labels = data.iloc[:, 0].values  # Labels (première colonne)

    # Normalisation des données
    std_scaler = StandardScaler()
    scaled_data = std_scaler.fit_transform(X)

    # Déterminer le nombre minimal de composantes pour 90% de variance expliquée
    pca_max = PCA(n_components=maxCol - 1)
    pca_max.fit(scaled_data)
    var_ratio = np.cumsum(pca_max.explained_variance_ratio_)

    nb_min_axes_90 = np.argmax(var_ratio > 0.9) + 1
    print(f"Le nombre minimum d'axes pour représenter 90% des données initiales est: {nb_min_axes_90}")

    # Appliquer l'ACP avec le bon nombre de composantes
    pca = PCA(n_components=nb_min_axes_90)
    pc = pca.fit(scaled_data)
    
    print(pca.explained_variance_)
    print(pca.explained_variance_ratio_)
    print(pca.components_)

    X_pca = pca.fit_transform(scaled_data)
  
    # Générer tous les scatter plots pour les paires de composantes principales
    pc_labels = [f"PC{i+1}" for i in range(nb_min_axes_90)]
    df_pca = pd.DataFrame(X_pca, columns=pc_labels)
    df_pca["Labels"] = labels  # Ajouter les labels pour annotation

    # Lister toutes les combinaisons de paires de composantes
    pc_pairs = [(i, j) for i in range(nb_min_axes_90) for j in range(i + 1, nb_min_axes_90)]

    # Tracer un cercle des corrélations pour chaque paire de composantes principales
    for i, j in pc_pairs:
        plot_correlation_circle(pc, data.columns[1:maxCol], pc_x=i, pc_y=j)
        
    # Ajuster la taille de la figure
    n_graphs = len(pc_pairs)
    fig, axes = plt.subplots(nrows=n_graphs // 2 + n_graphs % 2, ncols=2, figsize=(14, n_graphs * 3.5))
    axes = axes.flatten()
    
    for idx, (i, j) in enumerate(pc_pairs):
        ax = axes[idx]
        sns.scatterplot(data=df_pca, x=pc_labels[i], y=pc_labels[j], ax=ax)

        # Ajouter les annotations des points avec décalage pour éviter les superpositions
        for label, x, y in zip(labels, df_pca[pc_labels[i]], df_pca[pc_labels[j]]):
            ax.annotate(label, (x, y), textcoords="offset points", xytext=(5, 5), ha='left', fontsize=9)

        ax.set_xlabel(pc_labels[i])
        ax.set_ylabel(pc_labels[j])
        ax.set_title(f"{pc_labels[i]} vs {pc_labels[j]}")
        ax.grid()

    # Supprimer les sous-graphiques vides si le nombre de paires est impair
    for idx in range(len(pc_pairs), len(axes)):
        fig.delaxes(axes[idx])

    plt.tight_layout()
    plt.show()

```
