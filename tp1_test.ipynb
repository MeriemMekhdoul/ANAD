## Question 1
importation du dataFrame "Capital.xls"

```python
df = pandas.read_excel("Capital.xls")
df.head()
```

## Question 2:
Donner une représentation graphique en histogramme du pourcentage de villes par région


```python
df = df.rename(columns={"X": "Ville", "Y": "valeur", "Z": "Région"})

df.head()

#calculer le nombre de ville par région
#histogramme avec les region comme axe X et les pourcentages comme axe Y

nbVilles_par_region = df["Région"].value_counts()

pourcentage = nbVilles_par_region / nbVilles_par_region.sum() * 100

#mélanger les données ????
#pourcentage = pourcentage.sample(frac=1)
```

```python
# Tracer l'histogramme
plt.figure(figsize=(10, 6))
plt.bar(pourcentage.index, pourcentage.values, color="skyblue", width = 1.0)

# Ajouter titres et labels
plt.xlabel("Région")
plt.ylabel("Pourcentage de villes (%)")
plt.title("Histogramme du pourcentage de villes par région")

plt.show()
```

## Question 3

### Histogramme Horizontal

```python
plt.figure(figsize=(10, 6))

plt.barh(pourcentage.index, pourcentage.values, color="lightgreen")
plt.title("Histogramme horizontal du pourcentage de villes par région")
plt.xlabel("Pourcentage des villes (%)")
plt.ylabel("Région")

plt.show()
```

### Pie Chart

```python
plt.figure(figsize=(10, 6))

# Calculate the index of the largest slice
explode = [0] * len(pourcentage)  # Start with no explosions
largest_index = pourcentage.index(max(pourcentage))  # Find the index of the largest slice
explode[largest_index] = 0.1  # Explode the largest slice

plt.pie(pourcentage, labels = pourcentage.index)
plt.title("Diagramme circulaire du pourcentage de villes par région")

plt.show()
```
